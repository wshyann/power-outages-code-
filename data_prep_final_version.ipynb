## 2.2. Weather data

```python
carlisle_1
```

```python
# Carlisle weather data from 2020-01-01 to 2023-11-30
carlisle_3 = pd.read_csv("carlisle_2020-01-01_to_2023-11-30.csv")
```

```python
carlisle_3
```

```python
# All Weather data put together
carlisle_total = pd.concat([carlisle_1, carlisle_2, carlisle_3], ignore_index = True)
```

```python
carlisle_total.columns = carlisle_total.columns.str.capitalize()
```

```python
carlisle_total = carlisle_total.rename(columns = {"Name":"Location"})
```

### i. Dealing with missing values 

```python
carlisle_total.info()
```

```python
# Dropping categorical variables and probabilities
carlisle_total.drop(columns=["Stations", "Icon", "Conditions", "Preciptype", "Precipprob"], inplace = True)
carlisle_total
```

```python
carlisle_total[carlisle_total["Snow"].isna()].shape
```

```python
carlisle_total = carlisle_total[carlisle_total["Snow"].isna() != True]
carlisle_total
```

```python
carlisle_total[carlisle_total["Snowdepth"].isna()].shape
```

```python
carlisle_total["Windgust"] = carlisle_total["Windgust"].fillna(0)
carlisle_total
```

```python
carlisle_total.drop(columns = ["Severerisk", "Uvindex", "Solarenergy"], inplace = True)
```

# 3. Data selection and Pre-processing   

## 3.1.  Preparing outage data 

```python
# Removing all the exceptional events
hv_data_red2 = hv_data_red1
hv_data_red2
```

```python
hv_data_red2 = hv_data_red2.drop(columns = ["Exceptional Event ID", "Exceptional Event Start Date","Direct Cause", 
                                            "Exceptional Event EndDate", "Duration", "Location", "Year", "Direct Cause Category",
                                            "Customer Affected"])[(hv_data_red2["Year"]>=2010) & 
                                                                          (hv_data_red2["Location"] == "Carlisle") &
                                                                          (hv_data_red2["Direct Cause Category"] == "Weather and Environment")]
hv_data_red2
```

```python
hv_data_red2["Datetime"] = pd.to_datetime(hv_data_red2["Datetime"])
hv_data_red2
```

```python
# Rounding the datetime data to their nearest hours 

hv_data_red2["Datetime"] = hv_data_red2["Datetime"].dt.round("H")
```

```python
hv_data_red2["Outage"] = 1
```

```python
hv_data_red2.info()
```

## 3.1. Merging of data and preprocessing

```python
hv_data_red2
```

```python
carlisle_total
```

```python
carlisle_total["Datetime"] = pd.to_datetime(carlisle_total["Datetime"])
```

```python
merged_data = pd.merge(carlisle_total, hv_data_red2, on="Datetime", how="left")
```

```python
merged_data["Outage"] = merged_data["Outage"].fillna(0)
```

```python
merged_data.head()
```

```python
merged_data["Outage"] = merged_data["Outage"].astype("int")
```

```python
merged_data_final = merged_data.drop(columns = ["Datetime", "Location", "Feelslike", "Dew", "Visibility", "Snow"])
merged_data_final
```

```python
merged_data_final.info()
```

```python
merged_data_final = merged_data_final.dropna(subset=["Sealevelpressure"])
```

```python
merged_data_final.info()
```

```python
# This shows that we have a really unbalanced dataset
merged_data["Outage"].value_counts(normalize=True)
```

```python
merged_data_final.to_csv("merged_data_final.csv", index=False)
```
